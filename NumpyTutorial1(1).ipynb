```python
import numpy as np
array1= np.array([[1,1],[2,2]])
print(array1)
```

```python
array2random=np.random.rand(3,4)
print(array2random)
```

```python
array3=np.ones((3,3))
array4=np.zeros((3,3))
print("Array ones:\n", array3)
print("Array zeros:\n", array4)
```

```python
array5= np.random.rand(6,3)
print("Original array:\n", array5)
transpose_array5=array5.T
print("Transpose array:\n", transpose_array5)
```

```python
n=np.random.randint(2,8)
p=np.random.randint(2,8)
m=np.random.randint(2,8)

array_1=np.random.rand(n,p)
array_2=np.random.rand(p,m)

print("Matrix A:\n", array_1)
print("Matrix B:\n", array_2)

dot_product=np.dot(array_1,array_2)
print("\n Dot product:\n",dot_product)
```
